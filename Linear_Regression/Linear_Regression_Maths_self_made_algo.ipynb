# python class

we are going to create a class which works same as scikit learn regression class , with the help of maths we learned (open notebook)

```python
#we created functionn fit and predict instead of using scikit learn
class meraLR:
  def __init__(self):
    self.m =None
    self.b =None
  def fit(self,X_train,y_train):
    num=0
    den=0
    for i in  range (X_train.shape[0]):
      num=num+((X_train[i]-X_train.mean())*(y_train[i]-y_train.mean()))
      den=den+((X_train[i]-X_train.mean())*(X_train[i]-X_train.mean()))

    self.m=num/den
    self.b=y_train.mean()-(self.m*X_train.mean())
    print(self.m)
    print(self.b)




  def predict(self,X_test):
    print(X_test)
    return self.m*X_test + self.b
```
