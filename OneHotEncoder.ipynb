#
One Hot Encoding with Column Transformer 

```python
from sklearn.compose import ColumnTransformer
```

```python
transformer=ColumnTransformer(transformers=[("trnf1",OneHotEncoder(sparse_output=False,drop='first'),['fuel','owner'])])
```

```python
transformer.fit_transform(X_train)
```
