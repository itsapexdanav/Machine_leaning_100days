#Gradient Descent

Gradient Descent is an optimization algorithm used in machine learning and deep learning to minimize a loss (cost) function by iteratively updating model parameters.

```python
#making are own dataset and applying Gradient Descent
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)  #plottong scatter plot between X and y
```

#m and b using OLS

```python
#first we are using  ols so that we will get the value of m and b the
#we will verify through making constant slope and and b
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#drawing regression line
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color="red")
```

#Applying gradient descent by assuming slope is constant m=78.35
#And let's assume starting  value for intercept b=0

```python
y_pred=((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
#We are going to find intercept  with b=0(initially assume) and m=78.35
m=78.35#slope
b=0  #intercept
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# lets assume learning rate to be 0.1
rate=0.1
step_size=loss_slope*rate
step_size
```

```python
#calculating new intercept
b=b-step_size
print(b)
```

we have stareted with wrong value of b and moving toward correct value

```python
y_pred1=((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='black',label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

Blue line represents  the new b value as we have move toward correctt value of b     i.e., red

```python
#2nd iteration
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
step_size=loss_slope*rate
step_size
```

```python
#new value of b(intercept)
b=b-step_size
b
```

```python
#Again plotting  grapg and oberving the line get closer to correct line i.e, red
y_pred2=((78.35*X)+b).reshape(4)
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='yellow',label='b={}'.format(b))
plt.plot(X,y_pred1,color='black',label='b={}'.format(b))
plt.plot(X,y_pred,color='green',label='b=0')
plt.legend()
plt.show()
```

```python
#in above plot we can see thatline get closer to red ie, correct solution
```
