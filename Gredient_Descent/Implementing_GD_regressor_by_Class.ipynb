#Now Solving with our own Class (m and b both can vary)

```python
class GDRegressor:
  def __init__(self,learning_rate,epochs):
    self.m=100  #lets aasume initial value of m and b
    self.b=-120
    self.lr=learning_rate
    self.epochs=epochs

  def fit(self,X,y):
    #calculating b using GD
    for i in range(self.epochs):
      loss_slope_b=-2*np.sum(y-self.m*X.ravel()-self.b)
      #Slope with respect to b

      loss_slope_m=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())
      #slope wrt m
      self.b=self.b-(self.lr*loss_slope_b)
      self.m=self.m-(self.lr*loss_slope_m)
      # print("For i= ", {i},"  : ",loss_slope,self.b)
      # print("For i= ", {i},"  : ",loss_slope,self.b)

    print(self.m,self.b)

  def predict(self,X):
    return self.m*X+self.b






```

```python
gd=GDRegressor(0.001,50)   #(learning rate, epochs)
```

```python
gd.fit(X,y)
```
